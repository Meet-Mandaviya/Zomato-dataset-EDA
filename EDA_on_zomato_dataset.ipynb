# We used pandas merge function and parameter such as 'on' & 'how'
**on - on which feature we want to merge the datasets**

**how - which dataset we want to join with left or right**

```python
final_df = pd.merge(df,df_country,on='Country Code',how = 'left')
```

```python
final_df.head(2)
```

```python
final_df.dtypes
```

```python
final_df.columns
```

```python
country_names = final_df.Country.value_counts().index
```

```python
country_val = final_df.Country.value_counts().values
```

```python
# pie chart - top 3 countries that uses zomato
plt.pie(country_val[:3],labels = country_names[:3],autopct='%1.2f%%') # autopct='%1.2f%%'for percentage
```

**Observation: Zomato max records are from India
then usa, then uk**

```python
ratings = final_df.groupby(['Aggregate rating','Rating color','Rating text']).size().reset_index().rename(columns={0:'Rating count'})
```

```python
ratings
```

# **Observation:**
1. 4.5 to 4.9 --> Excellent
2. 4.0 to 4.4 --> Very good
3. 3.5 to 3.9 --> Good
4. 3.0 to 3.4 --> Average
5. 2.5 to 2.9 --> Average
6. 2.0 to 2.4 --> Poor

```python
ratings.head()
```

```python
import matplotlib
matplotlib.rcParams['figure.figsize'] = (12,6) # this line is to increase the size of the figure(it is most important)
sns.barplot(x = 'Aggregate rating',y = 'Rating count',data=ratings)
```

```python
sns.barplot(x = 'Aggregate rating',y = 'Rating count', hue = 'Rating color',data=ratings, palette = ['white','red','orange','yellow','green','green'])
```

# **Observations:**
1. Not rated count very high
2. max no of rating are between 2.5 and 3.5

# Count plot - plotting wrt to categorical variables x or y

```python
sns.countplot(x = 'Rating color',data=ratings,palette = ['white','red','orange','yellow','green','green'])
```
